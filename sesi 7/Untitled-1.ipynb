```python
# import package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# model Linear Regression
from sklearn.linear_model import LinearRegression
# code tambahan untuk membuat model Polynomial Regression. dengan code nya Linear Regression + ini = Polynomial Regrsssion
from sklearn.preprocessing import PolynomialFeatures

# untuk pisahin data jadi training dan testing data
from sklearn.model_selection import train_test_split
```

```python
df_bos = pd.read_csv("https://raw.githubusercontent.com/stefkwan-h8/dataset/main/boston.csv")
df_bos
```

```python
# merapikan data

# pilih fitur yang mau digunakan
# contoh kali ini, misal kita mau buat model yang bisa prediksi harga rumah dari jumlah kamarnya (kolom RM)
x = df_bos[["RM"]]
y = df_bos["MEDV"]

display(x.head(), y[:5])
```

```python
# cek missing values
df_bos.info()
```

```python
# sedikit eksplorasi data
df_bos.describe()
```

```python
# pisahin data jadi training dan testing data
x_train, x_test, y_train, y_test = train_test_split(x, y)

print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# buat dan latih model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# kalau ada 2 kamar, harga rumah berapa?
model.predict([[2]])
```

```python
# optional - visualisasi, karena cuma ada 1 fitur
plt.scatter(x=x, y=y)
plt.xlabel("jumlah kamar")
plt.ylabel("harga rumah")
plt.title("data harga rumah Boston")

y_pred = model.predict(x)
plt.plot(x, y_pred, color="orange")
plt.show()

```
