```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df_can = pd.read_excel("Canada.xlsx",
            sheet_name="Canada by Citizenship",
            skiprows=20,
            skipfooter=2)

display(df_can.head())
display(df_can.tail())
df_can.info()
```

```python
df_can_dropped = df_can.drop(columns=[ 'Type', 'Coverage',    'AREA', 'AreaName',      'REG',
        'RegName',      'DEV',  'DevName'])
```

```python
df_can_dropped.head()
```

```python
df_can_dropped = df_can_dropped.rename(columns={"OdName":"Country"})
df_can_dropped.head()
```

```python
df_can_dropped = df_can_dropped.set_index("Country")
df_can_dropped.head()
```

```python
df_indo = df_can_dropped.loc[["Indonesia"]]
df_indo
```

```python
df_indo.plot()
```

```python
df_indo.T.plot()
plt.title("Imigrasi Indonesia ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")
plt.show()
```

```python
df_indo.T.plot(figsize=(12,6), color="red", marker="o")
plt.title("Imigrasi Indonesia ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")
plt.show()
```

```python
df_indo.T.plot(figsize=(12,6), color="red", marker="o")
plt.title("Imigrasi Indonesia ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.text(2000.5, 1100,"Kerusuhan Mei 1998")

plt.show()
```

```python
df_mysg = df_can_dropped.loc[["Malaysia", "Singapore"]].T
df_mysg.head()
```

```python
df_mysg.plot()
```

```python
df_mysg.plot(figsize=(12,6), color=["blue", "green"], marker="o")
plt.title("Imigrasi Malaysia dan Singapore ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.text(1983, 2100, "1988 judicial crisis")

plt.show()
```

```python
df_mysg.plot(kind="area", figsize=(12,6), color=["blue", "green"])
plt.title("Imigrasi Malaysia dan Singapore ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.text(1983, 2100, "1988 judicial crisis")

plt.show()
```

```python
df_mysg.plot(kind="area", stacked=False, figsize=(12,6), color=["blue", "green"])
plt.title("Imigrasi Malaysia dan Singapore ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.text(1983, 2100, "1988 judicial crisis")

plt.show()
```

```python
df_mysg.plot(kind="bar", figsize=(12,6), color=["blue", "green"])
plt.title("Imigrasi Malaysia dan Singapore ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.text(4, 2100, "1988 judicial crisis")

plt.show()
```

```python
df_mysg.plot(kind="bar", figsize=(12,6), color=["blue", "green"])
plt.title("Imigrasi Malaysia dan Singapore ke Canada 1980-2013")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.annotate('1988 Judicial Crisis', 
             xy=(5.5, 1900),                  
             rotation=50,
             va='bottom',                  
             ha='left',)
plt.show()
```
